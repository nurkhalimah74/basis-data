#### 1.a. menghitung banyak total bilangan unik dalam list

```python
bilangan = [[1,2,3,4], [5,6,7,8], [2,4,1,7], [8,7], [9], [10]]
print ('total bilangan unik dalam list adalah ', len(bilangan))
```

```python
bilangan = [[1,2,3,4], [5,6,7,8], [2,4,1,7], [8,7], [9], [10]]
flatten_list = []
for sub_bilangan in bilangan:
    for item in sub_bilangan:
        flatten_list.append(item)
print(flatten_list)
print ('total bilangan unik dalam list adalah ', len(flatten_list))
```

#### 2. menghitung banyak mahasiswa basis data dan kalkulus

```python
keys =  [{
    'keys': 'basis data', 
            'nama mahasiswa': ['irvan', 'aji', 'ranu', 'lukman']}, 
        {
    'keys': 'kalkulus', 
            'nama mahasiswa': ['riga', 'sarah', 'irvan', 'aji']}]

for i in keys:
    mata_kuliah = i['keys']
    print(mata_kuliah.title())
    angka=0
    for nama in i['nama mahasiswa']:
        angka+=1
        print(str(angka) + '. ' + nama.title())
    print('banyak mahasiswa ', mata_kuliah, 'ada ', len(i['nama mahasiswa']), '\n')
    
basis_data = set(['irvan', 'aji', 'ranu', 'lukman'])
kalkulus = set(['riga', 'sarah', 'irvan', 'aji'])
matkul = basis_data & kalkulus
matkul_list = list(matkul)
print('mahasiswa basis data dan kalkulus adalah', matkul_list[1].title(), 'dan', matkul_list[0].title())
```

#### 5. apakah tahun yang diinput adalah tahun kabisat ?

```python
tahun = int(input('tahun berapa ? '))
if ((tahun %4 == 0 and tahun %100 != 0) or (tahun %400 == 0)):
    print('tahun ', tahun, ' adalah tahun kabisat')
else:
    print('tahun ', tahun, ' itu bukan tahun kabisat')
```

#### 6. input bilangan genap positif/negatif, ganjil positif/negatif, atau nol 

```python
A=[]
n=int(input("masukan banyaknya bilangan: "))

for i in range(n):
    a=int(input("masukan bilangan bulat: "))
    A.append(a)

print(' ')    
for j in A:
    if j==0:
        print("%d bukan bilangan positif maupun negatif" %j)
    elif j>0:
        if j%2 == 0:
            print("%d adalah bilangan genap positif" %j)
        else:
            print("%d adalah bilangan ganjil positif" %j)
    else:
        if j%2 == 0:
            print("%d adalah bilangan genap negatif" %j)
        else:
            print("%d adalah bilangan ganjil negatif" %j)
```
